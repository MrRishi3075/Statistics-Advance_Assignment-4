### Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

#### The t-test and z-test are both statistical methods used for hypothesis testing, but they differ in the situations in which they are applicable.

#### Z-Test:

* Use Case: The z-test is used when the population standard deviation is known, or the sample size is large (typically n > 30).

* Example Scenario: Suppose you want to test if the average height of a population is significantly different from a known value, and you have the population standard deviation available.

#### T-Test:

* Use Case: The t-test is used when the population standard deviation is unknown, and the sample size is small.

* Example Scenario: Consider a scenario where you want to test if there is a significant difference in the mean scores of two groups of students who received different teaching methods. If the sample size is small and the population standard deviation is unknown, you would use a t-test.

##### In summary, if you know the population standard deviation and/or have a large sample size, you may use a z-test. If the population standard deviation is unknown and you have a small sample size, a t-test is more appropriate.


### Q2: Differentiate between one-tailed and two-tailed tests.


### One-Tailed Test:

#### Definition: In a one-tailed (or one-sided) test, the critical region is on only one side of the distribution, either the right or the left.
* When to Use: When you are interested in detecting an effect in a specific direction (either an increase or a decrease).
* Example Scenario: Testing if a new drug increases the average reaction time of participants.

### Two-Tailed Test:

#### Definition: In a two-tailed (or two-sided) test, the critical region is on both sides of the distribution, allowing for the detection of effects in either direction.

* When to Use: When you are interested in detecting any significant difference, regardless of the direction (increase or decrease).
* Example Scenario: Testing if a new teaching method has a different effect on the mean scores of students compared to the current method.

##### In summary, the choice between one-tailed and two-tailed tests depends on the specific hypothesis you want to test and the direction of the effect you are interested in. One-tailed tests are more powerful when you have a specific directional hypothesis, while two-tailed tests are more conservative and suitable when you want to detect any significant difference, regardless of direction.

### Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.


### Type 1 Error (False Positive):

* Definition: Type 1 error occurs when you reject a null hypothesis that is actually true.

* Symbol: denoted by α (alpha).

* Example Scenario: Suppose you conduct a test to determine if a new drug is effective. A Type 1 error would be concluding that the drug is effective when, in reality, it has no effect.

### Type 2 Error (False Negative):

* Definition: Type 2 error occurs when you fail to reject a null hypothesis that is actually false.
* Symbol: denoted by β (beta).
* Example Scenario: Continuing with the drug example, a Type 2 error would be failing to detect that the drug is effective when, in reality, it does have a positive effect.

### Relationship between Type 1 and Type 2 Errors:

* There is a trade-off between Type 1 and Type 2 errors. As you decrease the risk of Type 1 error (e.g., by using a stricter significance level), the risk of Type 2 error tends to increase, and vice versa.

##### In summary, Type 1 error involves mistakenly concluding that there is an effect when there isn't, while Type 2 error involves failing to detect a real effect. The balance between these errors is essential in hypothesis testing, and it's often controlled through the choice of significance level (α) and statistical power.
